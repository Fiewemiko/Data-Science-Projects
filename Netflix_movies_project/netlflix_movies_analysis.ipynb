```python
import pandas as pd
import matplotlib.pyplot as plt

netflix = pd.read_csv("/Users/filip/Desktop/Python/nauka/Netflix_movies_project/netflix_data.csv")
#get rid of TV shows 
netflix_subset = netflix[netflix['type'] != "TV Show"] 
netflix_movies = netflix_subset[['title','country','genre','release_year',"duration"]]

shorter_movies = netflix_movies[netflix_movies['duration']< 60]

colors = []
for movie in netflix_movies['genre']:
    if movie == 'Children':
        colors.append('yellow')
    elif movie == 'Documentaries':
        colors.append('green')
    elif movie == 'Stand-Up':
        colors.append('blue')
    else:
        colors.append('red')


fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies['release_year'],netflix_movies['duration'],c=colors)
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie duration by Year of Release')

```
